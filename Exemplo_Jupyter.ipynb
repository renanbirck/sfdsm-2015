E também podemos plotar gráficos. Vejamos como isso será feito para um gráfico exemplo:

```python
fig = figure(figsize=(12, 9))  # Criar figura grande
years = [2010, 2011, 2012, 2013, 2014, 2015]  # Vetores de dados para exemplo
sales_A = [100, 150, 200, 300, 100, 150]
sales_B = [150, 220, 300, 330,  75,  55]

matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 


plot(years, sales_A, color='red', linestyle='-', marker='*', markerfacecolor='red')
plot(years, sales_B, color='black', linestyle='-',  marker='*', markerfacecolor='black')
xlabel('Ano', size=18)

# Definir limites dos eixos
xlim([2010, 2015])
ylim([50, 350])

ax = axes()
ax.set_xticklabels(years)
ylabel('Vendas (R$ mil)', size=16)
legend(['Carro A', 'Carro B'], fontsize=19, loc='best')
title('Vendas de Carros da Marca X', fontsize=16)
grid()
```
